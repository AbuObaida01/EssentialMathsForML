## Lecture 02 : Basics of Matrix Algebra


```python
p=np.array([[1,1],[5,-1],[3,2]])
print(p)
```

```python
q=np.array([[1,0],[6,-4],[3,8]])
print(q)
```

```python
r=np.add(p,q)
print(r)
```

```python
s=np.subtract(p,q)
print(s)
```

```python
t=np.array([[1,3,5,7],[8,4,9,10]])
print(t)
```

```python
u=np.dot(p,t)
print(u)
```

```python
w=np.array([[1,0,5],[6,-4,9],[3,8,8]])
print(w)
```

```python
det=np.linalg.det(w)
print(w)
```
